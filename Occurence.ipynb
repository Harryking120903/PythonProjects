```python
import numpy as np
import pandas as pd
import ast
```

```python
'''The aim of this file is to save the number of occurences of each word present in vocabulary in a spam or non-spam email.'''
df1=pd.read_csv('spam_ham_dataset.csv')
df1.drop("Unnamed: 0",axis=1)
df1.drop("label",axis=1)
print(df1.head())
```

```python
file=open('vocabulary.txt','r')
contents=file.read()
vocabulary=ast.literal_eval(contents)
```

```python
x=np.zeros((df1.shape[0],len(vocabulary)))
```

```python
y=np.zeros((df1.shape[0]))
```

```python
for i in range(df1.shape[0]):
    sub=df1.iloc[i,0].split()
    for subword in sub:
        if subword.lower() in vocabulary:
            x[i,vocabulary[subword]]+=1
            y[i]=df1.iloc[i,1]
np.save('x.npy',x)
np.save('y.npy',y)
```
