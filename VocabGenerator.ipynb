```python
import  pandas as pd
import nltk
from nltk.corpus import words
```

```python
'''In this file the aim is to generate the vocabulary for the classifier i.e. generate a set of words present in each subject.'''
vocabulary={}
df=pd.read_csv('spam_ham_dataset.csv')
df.drop("Unnamed: 0",axis=1)
df.drop("label",axis=1)
```

```python
df
```

```python
nltk.download('words')
set_words=set(words.words())
vocabulary={}
def build_vocab(curr_sub):
    idx=len(vocabulary)
    for word in curr_sub:
        if word.lower() not in vocabulary and word.lower() in set_words:'''checking if the word is a meaningful word.'''
            vocabulary[word]=idx
            idx+=1
if __name__=='__main__':
    for i in range(df.shape[0]):
        curr_sub=df.iloc[i,0].split()
        print(f'Current Email is {i}/{df.shape[0]}')
        print(f'Vocabulary length {len(vocabulary)}')
        build_vocab(curr_sub)
    fp=open('vocabulary.txt','w')
    fp.write(str(vocabulary))
    fp.close()
```
